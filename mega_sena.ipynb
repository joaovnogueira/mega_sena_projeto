## Juntando todas os números em uma só coluna


```python
bolas = ['Bola1', 'Bola2', 'Bola3', 'Bola4', 'Bola5', 'Bola6']
mega_sena['TodasBolas'] = mega_sena[bolas].astype(str).agg(','.join, axis=1)
mega_sena
```

```python
bolas_array = mega_sena['TodasBolas'].tolist()
bolas_array
```

## SEPARANDO TODOS OS NUMEROS DENTRO DE UM ARRAY

```python
array_resultado = []
for linha in mega_sena['TodasBolas']:
    array_resultado.extend(linha.split(','))  # Adicionar cada valor separado no array
```

```python
array_resultado
```

## Descobrindo os 6 valores mais frequentes na Mega-Sena


```python
top_6 = pd.Series(array_resultado).value_counts().head(6)
top_6
```

## Visualizando esses dados em um gráfico

```python
import matplotlib.pyplot as plt

top_6.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Top 6 Valores Mais Frequentes')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.xticks(rotation=0)
plt.show()
```

## Maior frequência de cada bola

```python
frequencia_cada_bola = []

for i in range(1, 7):
    bola_coluna = f'Bola{i}'

    # Calculando a frequência para cada bola
    frequencia = mega_sena[bola_coluna].mode()

    # Adicionando a frequência na lista
    frequencia_cada_bola.append(frequencia)
```

```python
frequencia_cada_bola
```

## Ano x quantidade de ganhadores


```python
ano_ganhadores = mega_sena.groupby('ano')['Ganhadores 6 acertos'].count()
ano_ganhadores
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ano_ganhadores.index, ano_ganhadores.values, marker='o', color='blue', linestyle='-', linewidth=2, markersize=6)

plt.title("Quantidade de Ganhadores de 6 Acertos da Mega-Sena por Ano")
plt.xlabel("Ano")
plt.ylabel("Número de Ganhadores")
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
